Simple RNN

```
import numpy as np
```

```
timesteps = 100     # Number of timesteps in the input
input_features = 32 # Dimensionality of the input feature space
output_features = 32 # Dimensionality of the output feature space
```

Get the input

```
inputs = np.random.random((timesteps, input_features))
```

Initial state is all zero vector

```
state_t = np.zeros((output_features))
```

Initialize weights randomely

```
W = np.random.random((output_features, input_features))
U = np.random.random((output_features, output_features))
b = np.random.random((output_features,) )
```

Let's implement RNN

```
succesive_outputs = []
for input_t in inputs:
  output_t = np.tanh(np.dot(W, input_t) + np.dot(U, state_t), + b)    # Combines input with the current state to obtain the current output
  succesive_outputs.append(output_t)
  state_t = output_t                # Update steps of the network for the next iteration

# The final output is a 2D tensor of shape(timesteps, output_features)
final_output_sequence = np.concatenate(succesive_outputs, axis=0)
```

**Issues**:



*   Too simplisctic for real lif problems
*   Not possible to learn long-term dependencies with simpleRNN due to vanishing gradient probelms: Adding more layers make model untrainable.
